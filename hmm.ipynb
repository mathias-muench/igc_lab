```python
import pandas as pd
import glob
from igc_lib.igc_lib import Flight
from pdflight import flight_to_dataframe,thermals_to_dataframe
flights = [Flight.create_from_file(i) for i in glob.glob("igc_files/575_MU.igc")]
[flight.glider_type for flight in flights]
#flight = Flight.create_from_file("/home/muenchm/thinclient_drives/C:/Users/muenchm/Desktop/2025-05-01 Wolfgang Janowitsch 562460.igc")
```

```python
fl = pd.concat(flight_to_dataframe(flight) for flight in flights)
fl
```

```python
import numpy as np
from hmmlearn import hmm
```

```python
X = fl.gsp.values.reshape(-1,1)
```

```python
# HMM mit 2 Zust√§nden (Gehen, Laufen)
model = hmm.GaussianHMM(n_components=3, covariance_type="diag", init_params="stmc")
model.startprob_ = [1.0, 0.0, 0.0]
model.fit(X)
display(model.means_)
display(model.covars_)
display(model.transmat_)
display(model.startprob_)
```

```python
# Viterbi-Algorithmus anwenden
zustandsfolge = model.predict(X)
zustandsfolge
```
